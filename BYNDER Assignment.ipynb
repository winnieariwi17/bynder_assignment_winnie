```python
import pandas as pd
import json
```

```python
assets_df= pd.read_csv('assets.csv')
companies_df=pd.read_csv('companies.csv')
users_df=pd.read_csv('users.csv')

#view of first few rows
print(assets_df.head())
print(companies_df.head())
print(users_df.head())
```

```python
#Drop PII data
users_df=users_df.drop(columns=['iban'])


# Split the name column into individual words, handle names with varying lengths
users_df['first_name'] = users_df['name'].str.split().str[:-1].str.join(' ')

# Drop the original 'name' column
users_df = users_df.drop(columns=['name'])

#Save the transformed data
users_df.to_csv('users_modified.csv', index=False)


print(users_df.head())
```

```python
#Standardize date format
users_df['created_at'].dtypes
assets_df['created_at'].dtypes
```

```python
#Convert date column to datetime
users_df['created_at']=pd.to_datetime(users_df['created_at'])
assets_df['created_at']=pd.to_datetime(assets_df['created_at'])

users_df['created_at'].dtypes
assets_df['created_at'].dtypes
```

```python
#Format all dates to YYYY-MM-DD
users_df['created_at'] = users_df['created_at'].dt.strftime('%Y-%m-%d')
assets_df['created_at'] = assets_df['created_at'].dt.strftime('%Y-%m-%d')
```

```python
import re

def standardize_columns(df):
    df.columns = df.columns.str.replace(r"([a-z])([A-Z])", r"\1_\2", regex=True)
    df.columns = df.columns.str.lower()  # Convert to lowercase
    df.columns = df.columns.str.replace(" ", "_")  # Replace spaces with underscores


# Apply the function to each DataFrame
standardize_columns(assets_df)
standardize_columns(companies_df)
standardize_columns(users_df)

#View of first five rows in standardized format
print(assets_df.head())
print(companies_df.head())
print(users_df.head())
```

```python
#Merge all CSV files
#Load the datasets
assets_df= pd.read_csv('assets.csv')
companies_df=pd.read_csv('companies.csv')
users_df=pd.read_csv('users_modified.csv')

#Apply the standardize_columns function to the DataFrames
standardize_columns(assets_df) # Standardize columns before merging
standardize_columns(companies_df)
standardize_columns(users_df)


#Merge Companies to Assets, using the standardized column name 'company_id'
assets_df=pd.merge(assets_df,companies_df,left_on='company_id',right_on='id')
assets_df=assets_df.drop(columns=['id_y']) #Drop the duplicate 'id' column


# Merge the new assets_df with users_df
all_data_df = pd.merge(assets_df, users_df, left_on='created_byuser_id', right_on='id')
all_data_df=all_data_df.drop(columns=['id']) #Drop the duplicate 'id' column

# Print or display the final merged DataFrame
print(all_data_df)
# or
display(all_data_df)



```

```python
#Remove duplicate company IDs
all_data_df.drop(columns=['company_id_x'], inplace= True)

print(all_data_df.head())
```

```python
#Rename columns
all_data_df=all_data_df.rename(columns= {'id_x':'asset_id','name_x':'asset_name','name_y':'company_name','company_id_y':'company_id','first_name':'username'})

print(all_data_df.head())
```

```python
#Upload JSON files
with open('asset_views.json','r') as f:
    asset_views_json=json.load(f)

with open('asset_downloads.json','r') as f:
    asset_downloads_json=json.load(f)

#Convert to PandasDF
asset_views_df=pd.DataFrame(asset_views_json)
asset_downloads_df=pd.DataFrame(asset_downloads_json)

#Print first rows of data
print(asset_views_df.head())
print(asset_downloads_df.head())
```

```python

#Standardize column names
def standardize_columns(df):
    df.columns = df.columns.str.replace(r"([a-z])([A-Z])", r"\1_\2", regex=True)
    df.columns = df.columns.str.lower()  # Convert to lowercase
    df.columns = df.columns.str.replace(" ", "_")  # Replace spaces with underscores

# Standardize column names for JSON DataFrames
standardize_columns(asset_views_df)
standardize_columns(asset_downloads_df)

#Print first rows of data
print(asset_views_df.head())
print(asset_downloads_df.head())

```

```python
asset_views_df['viewed_at'].dtypes
asset_downloads_df['downloaded_at'].dtypes
```

```python
#Convert to datetime
asset_views_df['viewed_at'] = pd.to_datetime (asset_views_df['viewed_at'],unit='ms')
asset_downloads_df['downloaded_at']= pd.to_datetime(asset_downloads_df['downloaded_at'],unit='ms')

#Standardize format to YYYY-MM-DD
asset_views_df['viewed at']= asset_views_df['viewed_at'].dt.strftime('%Y-%m-%d')
asset_downloads_df['downloaded_at']=asset_downloads_df['downloaded_at'].dt.strftime('%Y-%m-%d')


print(asset_views_df.head())
print(asset_downloads_df.head())
```

```python
#Merge asset views with all_data_df
merged_df=pd.merge(all_data_df,asset_views_df,left_on='created_byuser_id',right_on='unique_user_id')

print(merged_df.head())

print(merged_df.isnull())
```

```python
# Drop rows with any NaN values
merged_df.dropna(inplace=True)

print(merged_df.head())
```

```python

merged_df = merged_df.drop(columns=['unique_user_id'])

# Print the updated DataFrame to verify
print(merged_df.head())
```

```python
#Filter most viewed assets in 2024 for Reilly Limited.
reilly_2024_df = merged_df[
    (merged_df['company_name'] == 'Reilly Ltd') &
    (merged_df['viewed_at'].dt.year == 2024)
]

# Group by asset name and count views
asset_views_count = reilly_2024_df.groupby('asset_name')['asset_name'].count()

# 3. Sort by view count in descending order
most_viewed_assets = asset_views_count.sort_values(ascending=False)

# 4. Display the results
print(most_viewed_assets)
```

```python

df_to_download = merged_df

# Specify the file name and path
file_path = '/content/drive/MyDrive/merged_data.csv'

# Save the DataFrame to a CSV file
df_to_download.to_csv(file_path, index=False)

print(f"DataFrame saved to: {file_path}")
```
